```python
def main():
    try:
        results_listX = []
        results_listY = []
        while(1):
            try:
                case = int(input())
                if((case >= 1) & (case<=5)):
                    break
                else:
                    print("1에서 5 사이의 수를 입력하세요.")
            except ValueError:
                print("올바르지 않은 입력 형식입니다.")
        for i in range(case):
            xList = []
            yList = []
            xResult = 0
            yResult = 0
            for _ in range(3):
                while(1):
                    try:
                        x,y = input().split()
                        if((1<=int(x)&int(x)<=1000) & (1<=int(y)&int(y)<=1000)):
                            break
                        else:
                            print("1과 1000 사이의 숫자 두 개를 입력하세요")
                    except ValueError:
                        print("1과 1000 사이의 숫자 두 개를 입력하세요")
                    except UnboundLocalError:
                        print("올바르지 않은 입력 형식입니다.")
                xList.append(int(x))
                yList.append(int(y))
                
            vec01X = xList[1]-xList[0]
            vec10X = -1 * vec01X
            vec01Y = yList[1]-yList[0]
            vec10Y = -1 * vec01Y
            vec02X = xList[2]-xList[0]
            vec20X = -1*vec02X
            vec02Y = yList[2]-yList[0]
            vec20Y = -1*vec02Y
            vec12X = xList[2]-xList[1]
            vec21X = -1*vec12X
            vec12Y = yList[2]-yList[1]
            vec21Y = -1*vec12Y

            if(vec01X * vec02X + vec01Y*vec02Y == 0):
                xResult = xList[0]+vec01X+vec02X
                yResult = yList[0]+vec01Y+vec02Y
            if(vec10X * vec12X + vec10Y*vec12Y == 0):
                xResult = xList[1]+vec10X+vec12X
                yResult = yList[1]+vec10Y+vec12Y
            if(vec20X * vec21X + vec20Y*vec21Y == 0):
                xResult = xList[2]+vec20X+vec21X
                yResult = yList[2]+vec20Y+vec21Y

            if(xResult ==0 & yResult ==0):
                print("직사각형을 만들 수 없는 좌표입니다.")
            else:
                results_listX.append(xResult)
                results_listY.append(yResult)
        for i in range(case):
            print(results_listX[i],end=" ")
            print(results_listY[i])
    except KeyboardInterrupt:
        print("중단됨")
```
